```python
import pandas as pd
import numpy as np

print("Libraries pandas and numpy imported successfully.")
```

```python
num_entries = 100  # At least 50 entries

# Generate Customer_ID
customer_ids = np.arange(1, num_entries + 1)

# Generate Gender values
genders = np.random.choice(['Male', 'Female', 'Other'], size=num_entries)

# Generate Payment_Mode values
payment_modes = np.random.choice(['Credit Card', 'Debit Card', 'Net Banking', 'UPI', 'Cash on Delivery'], size=num_entries)

# Generate Product_Category values
product_categories = np.random.choice(['Electronics', 'Clothing', 'Home Goods', 'Books', 'Groceries', 'Beauty'], size=num_entries)

# Create a DataFrame
df = pd.DataFrame({
    'Customer_ID': customer_ids,
    'Gender': genders,
    'Payment_Mode': payment_modes,
    'Product_Category': product_categories
})

# Save the DataFrame to a CSV file
df.to_csv('customer_purchase_data.csv', index=False)

print(f"Synthetic dataset with {num_entries} entries generated and saved as 'customer_purchase_data.csv'.")
print("First 5 rows of the generated dataset:")
print(df.head())
```

```python
df_customers = pd.read_csv('customer_purchase_data.csv')
print("Dataset 'customer_purchase_data.csv' loaded successfully into df_customers.")
print("First 5 rows of the loaded dataset:")
print(df_customers.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate value counts for 'Product_Category'
product_category_counts = df_customers['Product_Category'].value_counts()

print("Value Counts for Product_Category:")
print(product_category_counts)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=product_category_counts.index, y=product_category_counts.values, palette='viridis')

# Set plot title and labels
plt.title('Distribution of Product Categories', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate value counts for 'Product_Category'
product_category_counts = df_customers['Product_Category'].value_counts()

print("Value Counts for Product_Category:")
print(product_category_counts)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=product_category_counts.index, y=product_category_counts.values, palette='viridis', hue=product_category_counts.index, legend=False)

# Set plot title and labels
plt.title('Distribution of Product Categories', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
gender_counts = df_customers['Gender'].value_counts()

print("Number of Customers per Gender:")
print(gender_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a cross-tabulation (contingency table) of Gender and Product_Category
gender_product_crosstab = pd.crosstab(df_customers['Gender'], df_customers['Product_Category'])

print("Cross-tabulation of Gender and Product_Category:")
print(gender_product_crosstab)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a stacked bar chart
gender_product_crosstab.plot(kind='bar', stacked=True, figsize=(12, 7), cmap='viridis')

# Set plot title and labels
plt.title('Product Category Distribution by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Purchases', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```
