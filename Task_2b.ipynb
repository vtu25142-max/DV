```python
import pandas as pd
import numpy as np

print("Libraries pandas and numpy imported successfully.")
```

```python
num_entries = 150

# 1. Generate Reading_ID
reading_ids = np.arange(1, num_entries + 1)

# 2. Create Time column (hourly readings over several days)
start_time = pd.to_datetime('2023-01-01 00:00:00')
time_stamps = pd.date_range(start=start_time, periods=num_entries, freq='H')

# 3. Generate Temperature_C (between 15 and 30 degrees Celsius)
temperature = np.random.uniform(15, 30, num_entries)

# 4. Generate Humidity_% (between 40 and 80 percent)
humidity = np.random.uniform(40, 80, num_entries)

# 5. Generate Air_Quality_Index (between 0 and 100)
air_quality = np.random.uniform(0, 100, num_entries)

# 6. Create a pandas DataFrame
data = {
    'Reading_ID': reading_ids,
    'Time': time_stamps,
    'Temperature_C': temperature,
    'Humidity_%': humidity,
    'Air_Quality_Index': air_quality
}
sensor_df = pd.DataFrame(data)

# 7. Save the DataFrame to a CSV file
csv_file_path = 'sensor_readings.csv'
sensor_df.to_csv(csv_file_path, index=False)

print(f"Synthetic dataset 'sensor_readings.csv' created with {num_entries} entries.")
print("First 5 rows of the generated dataset:")
print(sensor_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
num_entries = 150

# 1. Generate Reading_ID
reading_ids = np.arange(1, num_entries + 1)

# 2. Create Time column (hourly readings over several days)
start_time = pd.to_datetime('2023-01-01 00:00:00')
time_stamps = pd.date_range(start=start_time, periods=num_entries, freq='h')

# 3. Generate Temperature_C (between 15 and 30 degrees Celsius)
temperature = np.random.uniform(15, 30, num_entries)

# 4. Generate Humidity_% (between 40 and 80 percent)
humidity = np.random.uniform(40, 80, num_entries)

# 5. Generate Air_Quality_Index (between 0 and 100)
air_quality = np.random.uniform(0, 100, num_entries)

# 6. Create a pandas DataFrame
data = {
    'Reading_ID': reading_ids,
    'Time': time_stamps,
    'Temperature_C': temperature,
    'Humidity_%': humidity,
    'Air_Quality_Index': air_quality
}
sensor_df = pd.DataFrame(data)

# 7. Save the DataFrame to a CSV file
csv_file_path = 'sensor_readings.csv'
sensor_df.to_csv(csv_file_path, index=False)

print(f"Synthetic dataset 'sensor_readings.csv' created with {num_entries} entries.")
print("First 5 rows of the generated dataset:")
print(sensor_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
sensor_df = pd.read_csv('sensor_readings.csv')

print("First 5 rows of the sensor_df DataFrame:")
print(sensor_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nDataFrame Info:")
sensor_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries matplotlib.pyplot and seaborn imported successfully.")
```

```python
sensor_df['Time'] = pd.to_datetime(sensor_df['Time'])

# Create Temperature_Category
bins = [sensor_df['Temperature_C'].min() - 1, sensor_df['Temperature_C'].quantile(0.33), sensor_df['Temperature_C'].quantile(0.66), sensor_df['Temperature_C'].max() + 1]
labels = ['Low', 'Medium', 'High']
sensor_df['Temperature_Category'] = pd.cut(sensor_df['Temperature_C'], bins=bins, labels=labels, include_lowest=True)

# 1. Histogram for Temperature_C
plt.figure(figsize=(10, 6))
sns.histplot(data=sensor_df, x='Temperature_C', kde=True)
plt.title('Distribution of Temperature (Celsius)')
plt.xlabel('Temperature_C')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 2. KDE Plot for Humidity_%
plt.figure(figsize=(10, 6))
sns.kdeplot(data=sensor_df, x='Humidity_%', fill=True)
plt.title('Kernel Density Estimate of Humidity (%)')
plt.xlabel('Humidity_%')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 3. Rug Plot for Temperature_C
plt.figure(figsize=(10, 2))
sns.rugplot(data=sensor_df, x='Temperature_C')
plt.title('Rug Plot of Temperature (Celsius)')
plt.xlabel('Temperature_C')
plt.yticks([]) # Hide y-axis ticks for rug plot
plt.show()

# 4. Strip Plot for Air_Quality_Index
plt.figure(figsize=(10, 6))
sns.stripplot(data=sensor_df, x='Air_Quality_Index', jitter=True)
plt.title('Strip Plot of Air Quality Index')
plt.xlabel('Air_Quality_Index')
plt.ylabel('Observation')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 5. Swarm Plot for Temperature_C against Temperature_Category
plt.figure(figsize=(10, 6))
sns.swarmplot(data=sensor_df, x='Temperature_Category', y='Temperature_C')
plt.title('Swarm Plot of Temperature by Category')
plt.xlabel('Temperature Category')
plt.ylabel('Temperature_C')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 6. Line Plot for Time vs Temperature_C
plt.figure(figsize=(12, 6))
sns.lineplot(data=sensor_df, x='Time', y='Temperature_C')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature_C')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 7. Scatter Plot for Temperature_C vs Humidity_%
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sensor_df, x='Temperature_C', y='Humidity_%', hue='Air_Quality_Index', size='Air_Quality_Index', sizes=(20, 200), palette='viridis')
plt.title('Temperature vs Humidity (colored by Air Quality Index)')
plt.xlabel('Temperature_C')
plt.ylabel('Humidity_%')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Air Quality Index')
plt.show()
```

```python
print("Descriptive Statistics for Sensor Readings:")
print(sensor_df.describe().to_markdown(numalign="left", stralign="left"))

print("\nSkewness of Air_Quality_Index:")
print(sensor_df['Air_Quality_Index'].skew())

print("\nKurtosis of Air_Quality_Index:")
print(sensor_df['Air_Quality_Index'].kurtosis())
```
