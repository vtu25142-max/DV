```python
import pandas as pd
import numpy as np

# Create a synthetic dataset
np.random.seed(42) # for reproducibility

categories = ['Electronics', 'Clothing', 'Groceries', 'Home Goods', 'Books']
subcategories = {
    'Electronics': ['Mobile', 'Laptop', 'Headphones', 'Cameras'],
    'Clothing': ['T-Shirts', 'Jeans', 'Dresses', 'Jackets'],
    'Groceries': ['Fruits', 'Vegetables', 'Dairy', 'Snacks'],
    'Home Goods': ['Furniture', 'Kitchenware', 'Decor'],
    'Books': ['Fiction', 'Non-Fiction', 'Science', 'History']
}

data = []
num_rows = 50 # At least 30 rows

for _ in range(num_rows):
    category = np.random.choice(categories)
    subcategory = np.random.choice(subcategories[category])
    product = f'Product_{category[:3]}_{subcategory[:3]}_{np.random.randint(1, 100)}'
    sales_amount = np.random.randint(10, 1000)
    data.append([category, subcategory, product, sales_amount])

df = pd.DataFrame(data, columns=['Category', 'Subcategory', 'Product', 'Sales_Amount'])

# Save the dataset to a CSV file
df.to_csv('sales_data.csv', index=False)

print(f"Dataset created with {len(df)} rows and saved to 'sales_data.csv'")
```

```python
import plotly.express as px

# 3. Use Plotly to create a TreeMap
fig = px.treemap(
    df,
    path=[px.Constant("All"), 'Category', 'Subcategory', 'Product'],
    values='Sales_Amount',
    color='Category',
    hover_data=['Sales_Amount'],
    title='Sales Amount by Product Hierarchy (TreeMap)',
    color_discrete_sequence=px.colors.qualitative.Plotly # Changed to a different qualitative color sequence
)

fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
