```
import pandas as pd

# Load the Excel file into DataFrame
# Make sure to upload 'Maths.csv' (which appears to be an Excel file) using files.upload() first.
df = pd.read_excel("Maths.csv")

# Display first few rows
print("First few rows of the dataset:")
df.head()
```

```
print("Missing values in each column:")
df.isnull().sum()
```

```
df[['age', 'G1', 'G2', 'G3']].describe()
```

```
df['age'].unique()
```

```
# Fill missing values with mean (although df.isnull().sum() shows no missing values)
df['age'].fillna(df['age'].mean(), inplace=True)
df['G1'].fillna(df['G1'].mean(), inplace=True)
df['G2'].fillna(df['G2'].mean(), inplace=True)
df['G3'].fillna(df['G3'].mean(), inplace=True)

# Display updated DataFrame
df
```

```
df['Total_Score'] = df['G1'] + df['G2'] + df['G3']
df
```
