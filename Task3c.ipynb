```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Number of records
num_records = 150

# Generate data for each column
departments = ['HR', 'Sales', 'Marketing', 'Engineering', 'Finance']
genders = ['Male', 'Female', 'Non-binary']

data = {
    'Department': np.random.choice(departments, num_records),
    'Gender': np.random.choice(genders, num_records),
    'Salary': np.random.randint(40000, 120000, num_records), # Salaries between 40k and 120k
    'Satisfaction_Score': np.random.uniform(1.0, 5.0, num_records).round(1), # Scores between 1.0 and 5.0
    'Working_Hours': np.random.randint(30, 50, num_records), # Working hours between 30 and 50
    'Age': np.random.randint(22, 65, num_records) # Ages between 22 and 65
}

df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('employee_data.csv', index=False)

print('Dummy dataset "employee_data.csv" created successfully!')
print(f'Number of records: {len(df)}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

# 1. Bar Plot of average Salary across Department
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Salary', data=df, errorbar=None, palette='viridis')
plt.title('Average Salary Across Departments')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()

# 2. Box Plot of Satisfaction_Score across Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Satisfaction_Score', data=df, palette='pastel')
plt.title('Satisfaction Score Across Gender')
plt.xlabel('Gender')
plt.ylabel('Satisfaction Score')
plt.show()

# 3. Violin Plot of Working_Hours by Department
plt.figure(figsize=(12, 7))
sns.violinplot(x='Department', y='Working_Hours', data=df, inner='quartile', palette='muted')
plt.title('Distribution of Working Hours by Department')
plt.xlabel('Department')
plt.ylabel('Working Hours')
plt.show()

# 4. Ridgeline Plot of Salary grouped by Department (using seaborn.kdeplot in loop)
# Note: Ridgeline plots are not directly available in seaborn, but can be simulated using kdeplot.
# For a true ridgeline plot, libraries like joypy would be used, but require installation.
# We'll use a grouped KDE plot for a similar effect.

plt.figure(figsize=(12, 8))
for i, department in enumerate(df['Department'].unique()):
    sns.kdeplot(df[df['Department'] == department]['Salary'], fill=True, label=department, alpha=0.6)
plt.title('Kernel Density Estimate of Salary by Department')
plt.xlabel('Salary')
plt.ylabel('Density')
plt.legend(title='Department')
plt.show()

# 5. Beeswarm Plot of Satisfaction_Score by Gender using seaborn.swarmplot()
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Gender', y='Satisfaction_Score', data=df, palette='dark')
plt.title('Satisfaction Score Distribution by Gender (Beeswarm Plot)')
plt.xlabel('Gender')
plt.ylabel('Satisfaction Score')
plt.show()

# 6. Grouped Kernel Density Plot of Age grouped by Gender
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Age', hue='Gender', fill=True, common_norm=False, palette='coolwarm', alpha=0.5, linewidth=0)
plt.title('Kernel Density Estimate of Age by Gender')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
# Compute group-wise summary (mean, median) for Salary and Working_Hours by Department
print("\nGroup-wise Summary Statistics by Department:")
display(df.groupby('Department')[['Salary', 'Working_Hours']].agg(['mean', 'median']))

# Compute group-wise summary (mean, median) for Salary and Working_Hours by Gender
print("\nGroup-wise Summary Statistics by Gender:")
display(df.groupby('Gender')[['Salary', 'Working_Hours']].agg(['mean', 'median']))
```
